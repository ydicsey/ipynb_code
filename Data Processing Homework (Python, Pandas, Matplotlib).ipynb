### Compute the average payroll per year for all teams over 2011-2015, sort them in a descending order.

```python
mask1 = players['yearID'] <= 2015
mask2 = players['yearID'] >= 2011
# players.dropna(axis=0)
average_payroll = players[mask1 & mask2].pivot_table(index='teamID', columns='yearID', aggfunc='mean')['salary'].T.mean().reset_index().sort_values(by = 0, ascending = False).rename(columns={0: "average payroll"}).style.hide_index()
```

## task 2

### Compute the Batting Averages for each of the MLB teams over 2011-2015, sort them in a descending order. The Batting Average is defined as Hits/At Bats. The average is calculated from all players in each team.

```python
T_AB=players[mask1 & mask2].pivot_table(index='teamID', columns='yearID', aggfunc='sum')['AB'].T.sum().reset_index().sort_values(by = 0, ascending = False).rename(columns={0: "total_AB"}).style.hide_index()
T_H=players[mask1 & mask2].pivot_table(index='teamID', columns='yearID', aggfunc='sum')['H'].T.sum().reset_index().sort_values(by = 0, ascending = False).rename(columns={0: "total_H"}).style.hide_index()
```

```python
res = T_AB.data.join(T_H.data.set_index('teamID'), on='teamID')
res['Batting Average'] = res['total_H'].div(res['total_AB'])
res = total_win.data.join(res.set_index('teamID'), on='teamID')
res = res.sort_values('Batting Average', ascending=False)
res.plot('total_win', 'Batting Average', kind = 'scatter')
```
